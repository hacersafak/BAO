```python
# REAL DATA

import csv
import matplotlib.pyplot as plt
import numpy as np
import math
#REAL DATA WITH 1/4
ra=[]
dec=[]
z=[]
# Open the CSV file
with open(r"C:\Users\USER\Desktop\drpdatas\new-data\16761-around 0.9.csv", 'r') as csv_file:
    # Create a CSV reader
    csv_reader = csv.reader(csv_file)

    # Read the file row by row
    for row in csv_reader:
      ra.append(row[0])
      dec.append(row[1])
      z.append(row[2])
      
for i in range(len(ra)):
  ra[i]=float(ra[i])

for j in range(len(dec)):
  dec[j]=float(dec[j])

ra_deg = ra[:1000]
dec_deg = dec[:1000]

L = len(ra_deg)
ra_rad =  [math.radians(degree) for degree in ra_deg]
dec_rad = [math.radians(degree) for degree in dec_deg]

def angle_func(ra1_rad,dec1_rad,ra2_rad,dec2_rad):
  a= np.sin(dec1_rad)*np.sin(dec2_rad)+np.cos(dec1_rad)*np.cos(dec2_rad)*np.cos(ra1_rad-ra2_rad)
  delta=np.arccos(a)
  return delta

angles_DD = []
size = int((L*(L-1))/2)
angles_DD = [0] * size

step = 0
for i in range(L):
  for j in range(i+1,L):
    angles_DD[step] = angle_func(ra_rad[i],dec_rad[i],ra_rad[j],dec_rad[j] )
    step += 1

print(angles_DD[:10])

```

```python
#DD Histogram

plt.hist(angles_DD, bins=30,edgecolor="black")
```

```python
#Random data

ra_original=np.random.uniform(150,170,1000)
dec_original=np.random.uniform(0,20,1000)

ra_degree   = [float(entry) for entry in ra_original]
dec_degree   = [float(entry) for entry in dec_original]


ra_ran_rad = [math.radians(degree) for degree in ra_degree]
dec_ran_rad = [math.radians(degree) for degree in dec_degree]
```

```python
#RR Histogram

plt.hist(angles_RR, bins=30,edgecolor="black")
```

```python
#DR Calculation

angles_DR = []
size = int((L*(L-1))/2)
angles_DR = [0] * size

step = 0
for i in range(L):
  for j in range(i+1,L):
    angles_DR[step] = angle_func(ra_rad[i],dec_ran_rad[i],ra_rad[j],dec_ran_rad[j] )
    step += 1

print(angles_DR[:10])
```
