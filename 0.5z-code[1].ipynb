```python
#REAL DATA

import csv
import matplotlib.pyplot as plt
import numpy as np
import math
#REAL DATA WITH 1/32
ra=[]
dec=[]
z=[]
# Open the CSV file
with open(r"C:\Users\USER\Desktop\drpdatas\new-data\16757-around 0.5.csv", 'r') as csv_file:
    # Create a CSV reader
    csv_reader = csv.reader(csv_file)

    # Read the file row by row
    for row in csv_reader:
      ra.append(row[0])
      dec.append(row[1])
      z.append(row[2])
      
for i in range(len(ra)):
  ra[i]=float(ra[i])

for j in range(len(dec)):
  dec[j]=float(dec[j])

ra_deg = ra
dec_deg = dec

L = len(ra_deg)
ra_rad =  [math.radians(degree) for degree in ra_deg]
dec_rad = [math.radians(degree) for degree in dec_deg]

def angle_func(ra1_rad,dec1_rad,ra2_rad,dec2_rad):
  a= np.sin(dec1_rad)*np.sin(dec2_rad)+np.cos(dec1_rad)*np.cos(dec2_rad)*np.cos(ra1_rad-ra2_rad)
  delta=np.arccos(a)
  return delta

angles_DD = []
size = int((L*(L-1))/2)
angles_DD = [0] * size

step = 0
for i in range(L):
  for j in range(i+1,L):
    angles_DD[step] = angle_func(ra_rad[i],dec_rad[i],ra_rad[j],dec_rad[j] )
    step += 1

angles_DD.sort()
print(angles_DD[0])

#0.23960443823094074 biggest angle
#9.9491963474931e-05 smallest angle

```
